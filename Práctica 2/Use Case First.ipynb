# Carga de los datos

```python
base_path = "dataset/datos_hogar_2020.csv"
df = pd.read_csv(base_path)
df.head()
```

```python
counter = 0
for column in df:
    try:
        plt.figure()
        df.boxplot([column])
    except:
        pass
    
    plt.show()
    
    counter += 1
    if counter == 10:
        break

```

```python
def plot_distributions(dataframe, columns):
    for column in columns:
        data = dataframe[column]
        plt.figure()
        data = df[column]
        sns.displot(data, kde = True, bins = 100)
        plt.show()

plot_distributions(df, ["HB050", "DB100"])
```
